```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.iloc[:,1:]
```

```python
Y = df.iloc[:,0]
```

```python
X.shape
```

```python
Y.head()
```

```python
scaler = StandardScaler()
```

```python
scaledX = scaler.fit_transform(X)
```

```python
Xt = pd.DataFrame(scaledX, columns = X.columns)
```

```python
Xt.head()
```

We have transformed all the independent variables.

Now, lets see their skewness

```python
Xt.skew()
```

Since, outliers were inplace, the skewness is same as before. but normalisation took place within the data.

Just to deal with bias and variance when model training will be done, Normalization was necessary, to bring the data under same limits

This problem is a regression problem as the output variable is continous type of data

```python
for i in Xt.columns:
    plt.subplots()
    sns.distplot(Xt[i])
```

Every data point falls within the region of -3 to 3.

i.e, mean as 0 and standard deviation as 1

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(Xt,Y,test_size = 0.3, random_state = 33)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(xtrain,ytrain)
predlr = lr.predict(xtest)

from sklearn.metrics import r2_score, mean_squared_error

print(r2_score(ytest,predlr))
print(mean_squared_error(ytest,predlr))
```

Higher r2 score means higher predictability. The upper limit of r2 score is 1.

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr,xtrain,ytrain,cv=5,scoring='r2')

print(scores)
    
print('CV Mean: ', np.mean(scores))
```

There is a high difference between r2 and cross val r2

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()

rfr.fit(xtrain,ytrain)
predrf = rfr.predict(xtest)

print(r2_score(ytest,predrf))
print(mean_squared_error(ytest,predrf))
```

```python
scores = cross_val_score(rfr,xtrain,ytrain,cv=5,scoring='r2')

print(scores)
    
print('CV Mean: ', np.mean(scores))
```

we got here the max r2score as .88

Lets check with lasso and ridge regressions also.....

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1).fit(xtrain, ytrain)

predr = ridge.predict(xtest)

print(r2_score(ytest,predr))

scores1 = cross_val_score(ridge,xtrain,ytrain,cv=5,scoring='r2')

scores1.mean()
```

The ridge regression has pretty good r2 score. we can use this model as it has less difference with cross val score

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1).fit(xtrain, ytrain)

predl = lasso.predict(xtest)

print(r2_score(ytest,predl))

scores2 = cross_val_score(lasso,xtrain,ytrain,cv=5,scoring = 'r2')

scores2.mean()
```

The cross val score gives high score and the diff is also much. Lets tune the parameters for ridge and lasso before finalizing the model

```python
from sklearn.model_selection import GridSearchCV

alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(alpha=alpha)
grid = GridSearchCV(estimator=ridge, param_grid=param_grid, scoring='r2')
grid_result = grid.fit(xtrain, ytrain)
print('Best score: ', grid_result.best_score_)
print('Best params: ', grid_result.best_params_)
```

After tuning the paramter, the score is same as cross val score for ridge

```python
alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(alpha=alpha)
grid = GridSearchCV(estimator=lasso, param_grid=param_grid, scoring='r2')
grid_result = grid.fit(xtrain, ytrain)
print('Best score: ', grid_result.best_score_)
print('Best params: ', grid_result.best_params_)
```

The cross val score for lasso is same as score after tuning.

Since, the diff is low for ridge regression, we proceed with ridge

```python
ridge.score(xtest,ytest)
```

The model performs 77% accurately.
